```python
import qiskit
from qiskit.circuit import QuantumCircuit 
circuit=QuantumCircuit(6)
def conv_layer_1(circuit,unitary,param,n_qubits):
    qubit=[i for i in range(n_qubits)]
    if unitary=='simple_unitary':
        unitary=simple_unitary
    for i in range(3):
        unitary(circuit,param,qubit[2*i:2*i+2])
    for i in range(3):
        if i!=2:
            unitary(circuit,param,qubit[2*i+1:2*i+3])
        else:
            qubit=[5,0]
            unitary(circuit,param,qubit)
def conv_layer_2(circuit,unitary,param,n_qubits):
    qubit=[i for i in range(n_qubits)]
    if unitary=='simple_unitary':
        unitary=simple_unitary
    qubit=[0,2]
    unitary(circuit,param,qubit)
    qubit=[4,5]
    unitary(circuit,param,qubit)
    qubit=[2,4]
    unitary(circuit,param,qubit)
    qubit=[5,0]
    unitary(circuit,param,qubit)   

def conv_layer_3(circuit,unitary,param,n_qubits):
    qubit=[i for i in range(n_qubits)]
    if unitary=='simple_unitary':
        unitary=simple_unitary
        
    if unitary=='U_con3_qiskit':
        unitary=U_con3_qiskit
    qubit=[0,4]
    unitary(circuit,param,qubit)  

def simple_unitary(circuit,param,qubit):
    circuit.rx(param[0],qubit[0])
    circuit.rx(param[1],qubit[1])

def U_con3_qiskit(circuit,params,qubits):
    circuit.ry(params[0],qubits[0])
    circuit.ry(params[1],qubits[1])
    circuit.cx(qubits[1],qubits[0])
    circuit.ry(params[2],qubits[0])
    circuit.ry(params[3],qubits[1])
    circuit.cx(qubits[0],qubits[1])
    
conv_layer_1(circuit,U_con3_qiskit,[1,2,3,4],6)
conv_layer_2(circuit,U_con3_qiskit,[5,6,7,8],6)
conv_layer_3(circuit,U_con3_qiskit,[9,10,11,12],6)
# conv_layer_2(circuit,[3,4],1)

# from pylatexenc import latexwalker
# import pylatexenc
# # measrue 0,1,2
circuit.draw()
```
